# Ratio

```python
import pandas as pd
```

```python
df_p = pd.read_csv("/content/drive/MyDrive/data/processed/all_P.csv")
df_r = pd.read_csv("/content/drive/MyDrive/data/processed/all_R.csv")
```

```python
df_p=df_p.rename(columns={'Mainlabel':'label'})
df_r=df_r.rename(columns={'Mainlabel':'label'})
```

```python
df_p=df_p.drop(['Unnamed: 0','Start','End','Duration','Role'], axis=1)
df_r=df_r.drop(['Unnamed: 0','Start','End','Duration','Role'], axis=1)
```

```python
print (pd.DataFrame({'Percentage': df_p.groupby(('label')).size() / len(df_p)*100}))
```

```python
print (pd.DataFrame({'Percentage': df_r.groupby(('label')).size() / len(df_r)*100}))
```
